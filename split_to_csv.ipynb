```python
import numpy as np
import pandas as pd

file = pd.read_csv("players_20.csv")
```

```python
s = set()
for i in file["position"]:
    s.add(i)
s
```

```python
pd.concat([file.loc[file["position"] == "RES"], file.loc[file["position"] == "SUB"], file.loc[pd.isna(file["position"])]]).index
```

```python
nf = file.drop(index = pd.concat([file.loc[file["position"] == "RES"], file.loc[file["position"] == "SUB"], file.loc[pd.isna(file["position"])]]).index)
nf
```

```python
gk = file.loc[file["position"] == "GK",["id","diving","handling","kicking","reflexes","speed","positioning"]].reset_index()
gk
```

```python
gk.to_csv("goalkeepers.csv")
```
